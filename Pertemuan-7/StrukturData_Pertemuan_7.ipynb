Membuat Binary Tree Secara Manual

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

# Membuat tree secara manual
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

# Fungsi inorder traversal
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Membuat Binary Search Tree

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None


class BinarySearchTree:
  def __init__(self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

Traversal pada BST

```python
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

Pencarian Nilai dalam BST

```python
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)

# Uji pencarian
key = 60
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```

```python
# Program Binary Search Tree berdasarkan identitas saya

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def insert(self, root, data):

        # Jika node kosong, buat node baru
        if root is None:
            return Node(data)

        # Data lebih kecil masuk ke kiri
        if data < root.data:
            root.left = self.insert(root.left, data)

        # Data lebih besar masuk ke kanan
        else:
            root.right = self.insert(root.right, data)

        return root

# Fungsi traversal inorder
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)

# Fungsi traversal preorder
def preorder(node):
    if node:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)

# Fungsi traversal postorder
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")

# Fungsi pencarian data
def search(node, key):

    # Jika data ditemukan atau node kosong
    if node is None or node.data == key:
        return node

    # Cari ke kiri
    if key < node.data:
        return search(node.left, key)

    # Cari ke kanan
    return search(node.right, key)

# Data BST
data_list = [
    55,     # tiga digit terakhir NPM
    9,      # tanggal lahir
    10,     # bulan lahir
    65,     # ASCII huruf A
    117,    # ASCII huruf u
    7,      # angka bebas
    99      # angka bebas
]

# Membuat BST
bst = BinarySearchTree()
root = None

# Memasukkan data ke BST
for data in data_list:
    root = bst.insert(root, data)

# Menampilkan traversal
print("Inorder Traversal:")
inorder(root)

print("\n\nPreorder Traversal:")
preorder(root)

print("\n\nPostorder Traversal:")
postorder(root)

# Pencarian data
cari1 = 55
hasil1 = search(root, cari1)

print("\n")
if hasil1:
    print(cari1, "ditemukan dalam BST")
else:
    print(cari1, "tidak ditemukan dalam BST")

# Mencari data yang tidak ada
cari2 = 44
hasil2 = search(root, cari2)

if hasil2:
    print(cari2, "ditemukan dalam BST")
else:
    print(cari2, "tidak ditemukan dalam BST")
```
