Hapus Node

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add node in the beginning
    def AddBeginning(self, newdata):
        NewNode = Node(newdata)

        # Update new node next val to existing node
        NewNode.nextval = self.headval
        self.headval = NewNode

    # Function to remove node by key
    def RemoveNode(self, Removekey):
        HeadVal = self.headval

        if HeadVal is not None:
            if HeadVal.dataval == Removekey:
                self.headval = HeadVal.nextval
                HeadVal = None
                return

        while HeadVal is not None:
            if HeadVal.dataval == Removekey:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval

        if HeadVal is None:
            return

        prev.nextval = HeadVal.nextval
        HeadVal = None


# create linked list named Li
Li = LinkedList()

# add node
Li.AddBeginning("April")
Li.AddBeginning("Maret")
Li.AddBeginning("Febuari")

print("Sebelum dilakukan remove")
Li.listprint()

print("=================================")

# remove node by key "Maret"
Li.RemoveNode("Maret")

print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

Hapus di akhir

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add new node in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode

    # Function to remove last node
    def RemoveEnd(self):
        last = self.headval

        while last is not None:
            if last.nextval is None:
                break
            prev = last
            last = last.nextval

        if last is None:
            return

        prev.nextval = last.nextval
        last = None


# create linked list named Li
Li = LinkedList()

# add node
Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")

print("Sebelum dilakukan remove")
Li.listprint()

print("=================================")

# remove last node
Li.RemoveEnd()

print("Setelah dilakukan remove node terakhir")
Li.listprint()
```

Double Linked List

```python
# Class Node
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.prevval = None
        self.nextval = None

# Double LinkedList
class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

# Membuat 3 Node
dll = DoubleLinkedList()

n1 = Node("Senin")
n2 = Node("Selasa")
n3 = Node("Rabu")

dll.headval = n1

n1.nextval = n2
n2.prevval = n1
n2.nextval = n3
n3.prevval = n2

print("Data Awal:")
dll.listprint()

# Penyisipan di Tengah
def AddMiddle(self, middle_node, newdata):
    if middle_node is None:
        print("Node tidak ditemukan")
        return

    NewNode = Node(newdata)

    NewNode.nextval = middle_node.nextval
    middle_node.nextval = NewNode
    NewNode.prevval = middle_node

    if NewNode.nextval is not None:
        NewNode.nextval.prevval = NewNode

DoubleLinkedList.AddMiddle = AddMiddle

dll.AddMiddle(n2, "Kamis")

print("\nSetelah Penyisipan Tengah:")
dll.listprint()

# Hapus dengan Kata Kunci
def RemoveNode(self, key):
    head = self.headval

    while head is not None:
        if head.dataval == key:

            if head.prevval is not None:
                head.prevval.nextval = head.nextval
            else:
                self.headval = head.nextval

            if head.nextval is not None:
                head.nextval.prevval = head.prevval
            return

        head = head.nextval

DoubleLinkedList.RemoveNode = RemoveNode

dll.RemoveNode("Selasa")

print("\nSetelah Hapus 'Selasa':")
dll.listprint()
```

Circular Linked List

```python
# Class Node
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.prevval = None
        self.nextval = None

# Circular LinkedList
class CircularLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        if self.headval is None:
            return

        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

# Membuat 3 Node
cll = CircularLinkedList()

n1 = Node("Jumat")
n2 = Node("Sabtu")
n3 = Node("Minggu")

cll.headval = n1

n1.nextval = n2
n2.nextval = n3
n3.nextval = n1

print("Data Awal Circular:")
cll.listprint()

# Penyisipan di Akhir
def AddEnd(self, newdata):
    NewNode = Node(newdata)

    if self.headval is None:
        self.headval = NewNode
        NewNode.nextval = NewNode
        return

    last = self.headval
    while last.nextval != self.headval:
        last = last.nextval

    last.nextval = NewNode
    NewNode.nextval = self.headval

CircularLinkedList.AddEnd = AddEnd

cll.AddEnd("Senin")

print("\nSetelah Tambah di Akhir:")
cll.listprint()

# Hapus Node Terakhir
def RemoveEnd(self):
    if self.headval is None:
        return

    temp = self.headval

    if temp.nextval == self.headval:
        self.headval = None
        return

    prev = None
    while temp.nextval != self.headval:
        prev = temp
        temp = temp.nextval

    prev.nextval = self.headval

CircularLinkedList.RemoveEnd = RemoveEnd

cll.RemoveEnd()

print("\nSetelah Hapus Akhir:")
cll.listprint()
```
